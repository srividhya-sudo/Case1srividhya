```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
gaye =pd.read_csv('energy.csv')
print(gaye)
srh=gaye[gaye['State']=="Andhra"]
rcb=gaye[gaye['State']=="Karnataka"]
#LINE CHART ANDHRA KARNATAKA
fig1,diag1=plt.subplots(2)
diag1[0].plot(srh['Date'],srh['Consumption'])
diag1[1].plot(rcb['Date'],rcb['Consumption'],color='pink')
#bar chart Telangana Tripura
revanth=gaye[gaye['State']=='Telengana']
cm=gaye[gaye['State']=='Tripura']
fig,diag=plt.subplots(2)
diag[0].bar(revanth['Date'],revanth['Consumption'],color='green')
diag[1].bar(cm['Date'],cm['Consumption'],color='orange')
#pie chart
state1='Delhi'
state2='Odisha'
state3='Up'
state4='Punjab'
cse5=[state1,state2]
cse4=gaye[gaye['State']==state1]['Consumption'].sum()
cse3=gaye[gaye['State']==state2]['Consumption'].sum()
cse2=[cse3,cse4]
fig2,diag2=plt.subplots(2)
diag2[0].pie(cse2,labels=cse5,autopct='%1.1f%%')
cse5=[state3,state4]
cse4=gaye[gaye['State']==state3]['Consumption'].sum()
cse3=gaye[gaye['State']==state4]['Consumption'].sum()
cse2=[cse3,cse4]
diag2[1].pie(cse2,labels=cse5,autopct='%1.1f%%')
```
